## NumPy

[NumPy](https://numpy.org/devdocs/user/quickstart.html) 是一个用于处理数值计算的 Python 库，具有以下特点：

- **高性能多维数组对象**：提供快速、灵活的数组操作。
- **数学函数和运算**：包括三角函数、指数函数、逻辑运算等。
- **向量化操作**：一次性对整个数组进行操作，提高计算效率。
- **矩阵和线性代数支持**：进行矩阵运算、求解线性方程组等。
- **便捷的数据处理**：对数组进行切片、排序、聚合等操作。
- **与其他库的集成**：与诸如 Pandas、SciPy 等库配合使用。
- **广泛应用于科学计算**：在数据分析、机器学习、数值模拟等领域。
- **节省内存和提高速度**：有效地处理大型数据集。

## SciPy

SciPy 是一个基于 Python 的开源科学计算库，它建立在 NumPy 库的基础上，提供了许多高级的数值计算和科学计算工具。

- **丰富的功能**：包含各种数学、统计学、优化、信号处理等方面的函数和工具。
- **高效计算**：利用底层算法实现高效的数值计算。
- **与 NumPy 集成**：充分利用 NumPy 的数组功能。
- **优化算法**：例如最小化函数、方程求解等。
- **信号处理**：滤波、频谱分析等。
- **统计分析**：假设检验、方差分析等。
- **图像处理**：图像增强、变换等。
- **集成其他库**：可与其他科学计算库协同工作。
- **广泛应用领域**：适用于物理学、工程学、计算机科学等多个领域。

## Pandas

- **数据读取和写入**：支持多种数据格式的读取和写入。
- **数据清洗和预处理**：方便进行缺失值处理、重复值删除等。
- **数据选择和过滤**：基于条件选择数据。
- **数据聚合和分组**：进行聚合计算和分组操作。
- **数据排序和排名**：按指定列进行排序。
- **数据合并和连接**：方便地合并多个数据集。
- **时间序列处理**：专门的函数处理时间序列数据。
- **数据可视化**：与其他可视化库集成。

## Matplotlib

Matplotlib 是一个广泛使用的 Python 绘图库，用于创建各种类型的可视化图形，具有以下特点：
- **丰富的绘图类型**：包括线图、散点图、柱状图、饼图等。
- **灵活定制**：可以自定义图形的各种属性，如颜色、标签、线条样式等。
- **可交互性**：支持鼠标交互和图形的动态更新。
- **高质量输出**：生成高质量的图形，可用于报告、论文等。
- **刻度和标签控制**：方便地设置坐标轴刻度和标签。
- **多轴绘图**：在一个图形中绘制多个子图。
- **图例和注释**：添加图例和注释来解释图形。
- **数据可视化**：帮助理解和分析数据。


```python
from IPython.display import IFrame
IFrame('cheatsheet/Matplotlib.pdf', width = 1000, height = 600)
```

```python
import numpy as np
import matplotlib.pyplot as plt

r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r
fig, ax = plt.subplots(
  subplot_kw = {'projection': 'polar'} 
)
ax.plot(theta, r)
ax.set_rticks([0.5, 1, 1.5, 2])
ax.grid(True)
plt.show()
```

以下是浙江省2000年至2010年的人口变化表格：
|年份|总人口数（万人）|
|:--:|:--:|
|2000|4501.22|
|2001|4519.84|
|2002|4535.98|
|2003|4551.58|
|2004|4577.22|
|2005|4602.11|
|2006|4629.43|
|2007|4659.34|
|2008|4687.85|
|2009|4716.18|
|2010|4748.05|

画图显示其变化趋势。

```python
import numpy as np
import matplotlib.pyplot as plt

year = np.arange(2000, 2011)
population = [4501.22, 4519.84, 4535.98, 4551.58, 4577.22, 4602.11, 4629.43, 4659.34, 4687.85, 4716.18, 4748.05]

plt.plot(year, population, 'b.', label='population')
plt.xlabel('year')
plt.ylabel('population')
plt.legend()
plt.show()
```
